```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.utils import shuffle
from sklearn.metrics import accuracy_score
```

```python
url='toy.csv'
df = pd.read_csv(url)
print(df)
```

```python
y = df.category
X = df.words

df['words'].value_counts()
```

```python
le=LabelEncoder()
le.fit(df['category'].drop_duplicates(keep='first'))

y_encoder=[]
y_encoder=le.transform(y)


X_encoder=[]
vectorizer = CountVectorizer()
X_encoder = vectorizer.fit_transform(X)
```

```python
print(X[0])
print(X_encoder[0])
print(vectorizer.inverse_transform(X_encoder[0]))
print(y_encoder)
```

```python
n#ew_X = []
#for x in X_encoder: new_X.append(list(x.toarray())[0])
X_train, X_test, y_train, y_test = train_test_split(X_encoder.toarray() , y_encoder, test_size=0.3, random_state=42)
print(X_train)
#X_encoder_array = X_encoder.toarray()
#total = len(X_encoder_array)
#X_train, X_test, y_train, y_test = X_encoder_array[:total*3/10], X_encoder_array[total*3/10:], y_train[:total*3/10], y_test[total*3/10:]
```

```python
print(X_train.toarray())
```

```python
clf = GaussianNB()
clf.fit(X_train, y_train)
y_predict=clf.predict(X_test)
accuracy_score(y_test, y_predict, normalize=False)
```
