```python
import numpy as np
import jieba
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.utils import shuffle
from sklearn.metrics import accuracy_score
import glob, os
```

```python
dfs = []
for file_num in range(0, 26):
    url = "data"+str(file_num)+"_char.csv"
    t = pd.read_csv(url, index_col=None, header=0)
    dfs.append(t)
df  = pd.concat(dfs, axis=0, ignore_index=True)
df=shuffle(df)
df.count()
```

```python
y = df.category
X = [str(w) for w in df.words]
```

```python
le=LabelEncoder()
le.fit(df['category'].drop_duplicates(keep='first'))

y_encoder=le.transform(y)

vectorizer = CountVectorizer(analyzer='char')
X_encoder = vectorizer.fit_transform(X)
```

```python
print(X[0])
print(X_encoder[0])
print(vectorizer.inverse_transform(X_encoder[0]))
print(y_encoder)
```

```python
# X_train, X_test, y_train, y_test = train_test_split(X_encoder.toarray() , y_encoder, test_size=0.3, random_state=42)
test_ratio=0.3
test_size=int (test_ratio*df['category'].count())
X_train, X_test, y_train, y_test = X_encoder.toarray()[test_size:], X_encoder.toarray()[:test_size], y_encoder[test_size:], y_encoder[:test_size] 
```

```python
clf = GaussianNB()
clf.fit(X_train, y_train)
```

```python
y_predict=clf.predict(X_test)
accuracy_score(y_test, y_predict)
```
